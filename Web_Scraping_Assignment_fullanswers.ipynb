## 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

## 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
imdb_page = requests.get("https://www.imdb.com/list/ls091520106/?sort=list_order,asc&st_dt=&mode=simple&page=1&ref_=ttls_vw_smp")

imdb = BeautifulSoup(imdb_page.content)

imdb_movies = imdb.find_all("span", class_="lister-item-header")
movie_names = []
for i in imdb_movies:
    for j in i.find_all("span"):
        movie_names.append(j.text.replace("\n",""))
name = []
for i in range(1,len(movie_names),3):
    name.append(movie_names[i])
    
year = []
for i in range(2,len(movie_names),3):
    year.append(movie_names[i])
    
imdb_rating = imdb.find_all("div", class_="col-imdb-rating")
rating = []
for i in imdb_rating:
    for j in i.find_all("strong"):
        rating.append(j.text.replace("\n","").strip())


df_imdb=pd.DataFrame({})
df_imdb['Movie name']=name
df_imdb['Rating']=rating
df_imdb['Release year']=year
df_imdb
```

### i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

## 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. ## 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
page=requests.get("https://internshala.com/fresher-jobs")

soup=BeautifulSoup(page.content)

job_title=soup.find_all("div", class_="heading_4_5 profile")
j_title=[]
for i in job_title:
    j_title.append(i.text.replace("\n","").strip())

company_names=soup.find_all("div", class_="heading_6 company_name")
comp_name=[]
for i in company_names:
    comp_name.append(i.text.replace("\n","").strip())

CTC_l=soup.find_all("div", class_="item_body")
CTC=[]
for i in CTC_l:
    for j in i.find_all("i"):
        CTC.append(i.text.replace("\n","").strip())

apply=soup.find_all("div", class_="item_body")
apply_date=[]
for i in apply:
    apply_date.append(i.text)
appndates=[]
for i in range(2, len(apply_date), 3):
    appndates.append(apply_date[i])


df=pd.DataFrame({})
df['Job Title']=j_title
df['Company Name']=comp_name
df['CTC']=CTC
df['Apply Date']=appndates
df
```
